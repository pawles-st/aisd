```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
data_begin = 1000                                           # beginning of data range
data_end = 100000                                           # end of data range (inclusive)
data_step = 1000                                            # data range step
data_range = np.arange(data_begin, data_end + 1, data_step) # array of data values
data_size = len(data_range)                                 # length of the data values array

element_types = ["begin", "middle", "end", "nonexistent", "random"]

m = 100 # number of repeats for a single experiment

comparisons = np.zeros((len(element_types), data_size), int)
time = np.zeros((len(element_types), data_size), int)

for t in range(len(element_types)): # element type

    for n in data_range: # size of the input

        # read results and sum them

        data = open(f"results/search_{element_types[t]}_{n}.txt", "r")
        for result in data.readlines():
            result_split = result.rstrip("\n").split(" ")
            result_split = [eval(i) for i in result_split]
            comparisons[t][n // data_step - 1] += result_split[0]
            time[t][n // data_step - 1] += result_split[1]

    # take the average

    comparisons[t] = comparisons[t] / m
    time[t] = time[t] / m

# plot the averages

for t in range(0, len(element_types)):
    plt.plot(data_range, comparisons[t], label=f"{element_types[t]}")
    
plt.title("binary search comparisons")
plt.legend()
plt.show()

for t in range(len(element_types)):
    plt.plot(data_range, time[t], label=f"{element_types[t]}")
    
plt.title("binary search time")
plt.legend()
plt.show()
```
