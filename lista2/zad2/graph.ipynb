```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
N_values = np.arange(10, 201, 10)

for k in [1, 10, 100]:
    
    for t in ["random", "desc", "asc"]:

        INS_c = np.zeros(20, int)
        INS_s = np.zeros(20, int)
        MER_c = np.zeros(20, int)
        MER_s = np.zeros(20, int)
        QUI_c = np.zeros(20, int)
        QUI_s = np.zeros(20, int)

        for N in N_values:

            # read results

            ins = open(f"results/insertion_sort_{t}_{k}_{N}.txt", "r")
            for result in ins.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                INS_c[N // 10 - 1] += result_split[0]
                INS_s[N // 10 - 1] += result_split[1]

            mer = open(f"results/merge_sort_{t}_{k}_{N}.txt", "r")
            for result in mer.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                MER_c[N // 10 - 1] += result_split[0]
                MER_s[N // 10 - 1] += result_split[1]

            qui = open(f"results/quick_sort_{t}_{k}_{N}.txt", "r")
            for result in qui.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                QUI_c[N // 10 - 1] += result_split[0]
                QUI_s[N // 10 - 1] += result_split[1]
        
        INS_c = INS_c / k
        INS_s = INS_s / k
        MER_c = MER_c / k
        MER_s = MER_s / k
        QUI_c = QUI_c / k
        QUI_s = QUI_s / k

        plt.plot(N_values, INS_c, label="insert sort")
        plt.plot(N_values, MER_c, label="merge sort")
        plt.plot(N_values, QUI_c, label="quick sort")
        plt.title(f"comparisons, type={t}, k={k}")
        plt.legend()
        plt.show()
        plt.plot(N_values, INS_s, label="insert sort")
        plt.plot(N_values, MER_s, label="merge sort")
        plt.plot(N_values, QUI_s, label="quick sort")
        plt.title(f"replacements, type={t}, k={k}")
        plt.legend()
        plt.show()
        
        plt.plot(N_values, INS_c / N_values, label="insert sort")
        plt.plot(N_values, MER_c / N_values, label="merge sort")
        plt.plot(N_values, QUI_c / N_values, label="quick sort")
        plt.title(f"comparisons / n, type={t}, k={k}")
        plt.legend()
        plt.show()
        plt.plot(N_values, INS_s / N_values, label="insert sort")
        plt.plot(N_values, MER_s / N_values, label="merge sort")
        plt.plot(N_values, QUI_s / N_values, label="quick sort")
        plt.title(f"replacements / n, type={t}, k={k}")
        plt.legend()
        plt.show()
    
```
