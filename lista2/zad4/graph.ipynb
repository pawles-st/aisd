```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
N_values = np.arange(10, 201, 10)

for k in [1, 10, 100]:
    
    for t in ["random", "desc", "asc"]:

        INS_c = np.zeros(20, int)
        INS_s = np.zeros(20, int)
        QUI_c = np.zeros(20, int)
        QUI_s = np.zeros(20, int)
        HYB_c = np.zeros(20, int)
        HYB_s = np.zeros(20, int)

        for N in N_values:

            # read results

            ins = open(f"../zad2/results/insertion_sort_{t}_{k}_{N}.txt", "r")
            for result in ins.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                INS_c[N // 10 - 1] += result_split[0]
                INS_s[N // 10 - 1] += result_split[1]

            qui = open(f"../zad2/results/quick_sort_{t}_{k}_{N}.txt", "r")
            for result in qui.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                QUI_c[N // 10 - 1] += result_split[0]
                QUI_s[N // 10 - 1] += result_split[1]
                
            hyb = open(f"results/hybrid_{t}_{k}_{N}.txt", "r")
            for result in hyb.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                HYB_c[N // 10 - 1] += result_split[0]
                HYB_s[N // 10 - 1] += result_split[1]
        
        INS_c = INS_c / k
        INS_s = INS_s / k
        QUI_c = QUI_c / k
        QUI_s = QUI_s / k
        HYB_c = HYB_c / k
        HYB_s = HYB_s / k

        plt.plot(N_values, INS_c, label="insert sort")
        plt.plot(N_values, QUI_c, label="quick sort")
        plt.plot(N_values, HYB_c, label="hybrid sort")
        plt.title(f"comparisons, type={t}, k={k}")
        plt.legend()
        plt.show()
        plt.plot(N_values, INS_s, label="insert sort")
        plt.plot(N_values, QUI_s, label="quick sort")
        plt.plot(N_values, HYB_s, label="hybrid sort")
        plt.title(f"replacements, type={t}, k={k}")
        plt.legend()
        plt.show()
    
```

```python
N_values = np.arange(1000, 20001, 1000)

for k in [1, 10, 100]:
    
    for t in ["random", "desc", "asc"]:

        QUI_c = np.zeros(20, int)
        QUI_s = np.zeros(20, int)
        DUAL_c = np.zeros(20, int)
        DUAL_s = np.zeros(20, int)

        for N in N_values:

            # read results
            
            qui = open(f"../zad2/results/quick_sort_{t}_{k}_{N}.txt", "r")
            for result in qui.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                QUI_c[N // 1000 - 1] += result_split[0]
                QUI_s[N // 1000 - 1] += result_split[1]

            dual = open(f"results/dual_quick_sort_{t}_{k}_{N}.txt", "r")
            for result in dual.readlines():
                result_split = result.rstrip("\n").split(" ")
                result_split = [eval(i) for i in result_split]
                DUAL_c[N // 1000 - 1] += result_split[0]
                DUAL_s[N // 1000 - 1] += result_split[1]
        
        QUI_c = QUI_c / k
        QUI_s = QUI_s / k
        DUAL_c = DUAL_c / k
        DUAL_s = DUAL_s / k
        
        #plt.plot(N_values, QUI_c  / N_values / np.log(N_values), label="quick sort")
        plt.plot(N_values, DUAL_c / N_values / np.log(N_values), label="dual quick sort")
        plt.title(f"comparisons / nlog(n), type={t}, k={k}")
        plt.legend()
        plt.show()

        plt.plot(N_values, QUI_c, label="quick sort")
        plt.plot(N_values, DUAL_c, label="dual quick sort")
        plt.title(f"comparisons, type={t}, k={k}")
        plt.legend()
        plt.show()
        plt.plot(N_values, QUI_s, label="quick sort")
        plt.plot(N_values, DUAL_s, label="dual quick sort")
        plt.title(f"replacements, type={t}, k={k}")
        plt.legend()
        plt.show()
        
        plt.plot(N_values, QUI_c / N_values, label="quick sort")
        plt.plot(N_values, DUAL_c / N_values, label="dual quick sort")
        plt.title(f"comparisons / n, type={t}, k={k}")
        plt.legend()
        plt.show()
        plt.plot(N_values, QUI_s / N_values, label="quick sort")
        plt.plot(N_values, DUAL_s / N_values, label="dual quick sort")
        plt.title(f"replacements / n, type={t}, k={k}")
        plt.legend()
        plt.show()
    
```
